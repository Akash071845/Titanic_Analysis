```python
import pandas as pd
df = pd.read_csv("titanic.csv")
print(df)
```

```python
# this is the dataset of showing the passengers info in the titanic
```

```python
print(df.head(10))
```

```python
passengers_survived = df['survived'].value_counts()
print(passengers_survived)
```

```python
# this shows the no of passengers survived and no of passengers who perished in the titanic accident 
```

```python
passengers_older_than_30 = df[df['age'] >= 30]
print(passengers_older_than_30)
```

```python
# this shows the passengers who are more than or equal to 30 years who travelled in the titanic
```

```python

no_of_people_in_each_class = df['pclass'].value_counts()

print(no_of_people_in_each_class)
```

```python
#this shows the number of people in each class as there was discrimination based on class during the sail of the titanic
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

sns.countplot(data=df, x='sex', hue='survived')

plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel("No of Passengers")

plt.legend(title='survived', labels=['No', 'Yes'])

plt.show
```

```python
# this graph shows that more female passengers survived compared to male passengers in titanic incident
```

```python
sns.countplot(data=df, x='pclass', hue='survived') # hue is used to break the survived categ into two more types survived and non survived
plt.title('Survival Count by Passnger Class')

plt.xlabel('Class')
plt.ylabel('No of Passengers')

plt.legend(title="survived", labels=['No', 'Yes'])

plt.show()
```

```python
# this plot shows that the passengers from the third class suffered heavy loss of life
```

```python
sns.boxplot(data=df, x='pclass', y='fare', hue='pclass', palette='Pastel1', legend=False)

plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare Distribution')

plt.show()
```

```python
# this plot represents that the passengers from the first class paid expensive fair for a ride in the titanic
```

```python
sns.histplot(data=df, x='age', bins=30, kde=True, color='green')

plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# so people in the age between 20-30 have travelled more in the titanic
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Between Features")
plt.show()
```
